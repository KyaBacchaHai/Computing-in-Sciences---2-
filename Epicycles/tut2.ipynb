# Task 1:

```python
def fourier_coeff(zvals, m):    
    numPoints = shape(zvals)[0]  
    tvals = linspace(0,1,numPoints+1)[1:]    
    c_m = 1.0/numPoints * np.sum( zvals * np.exp(-2*pi*1j*m*tvals) )    
    return c_m
```

```python
def compute_Fourier_dict( zvals, n_max ):
    req_dict = {}
    for m in range(-1*n_max,n_max+1):
        cm = fourier_coeff(zvals[:-1],m)
        req_dict[m] = cm
    return req_dict;
    
```

# Task 2:

```python
def reconstruct_Fourier_dict( fourier_dict ):
    z_vals = zeros(100, dtype = complex)
    tvals= linspace(0,1,101)[1:]
    for key in fourier_dict.keys():
        z_vals += fourier_dict[key] * np.exp(2*pi*1j*key*tvals)
    plot(z_vals.real, z_vals.imag, ".")
    xlabel("m -->")
    ylabel("Real ( C(m) ) -->")
```
