# Prey Predator Model

```python
# initialie parameters alpha, beta, gamma, delta
params = (2.,1.,1.,2.)
```

```python
def rate(state, t=None, cons=params):
    return [(cons[0]-cons[1]*state[1])*state[0],(cons[2]*state[0]-cons[3])*state[1]]
```

```python
state = [1.,1.]
dt = 0.01
for t in range(1000):
    [x, y] = rate(state)
    arrow(state[0], state[1], x*dt, y*dt)
    state[0]+=x*dt
    state[1]+=y*dt
    ylim(0, 5)
    xlim(0, 5)
```

### Finding the steady state

```python
steadystate = opt.root(rate, [5,5], ()).x
```

```python
print(steadystate)
```
